## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

# Create a new column 'line_subtotal' by multiplying 'unit_price' and 'qty'
df['line_subtotal'] = df['unit_price'] * df['qty']

# Print only the specified columns without the index
df[['unit_price', 'qty', 'line_subtotal']].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# Calculate total weight for each row
df['total_weight'] = df['unit_weight'] * df['qty']

# Define conditions for shipping price
conditions = [
    (df['total_weight'] > 50),
    (df['total_weight'] <= 50)
]

# Define corresponding shipping prices
choices = [
    df['total_weight'] * 7,
    df['total_weight'] * 10
]

# Create the 'shipping_price' column based on conditions
df['shipping_price'] = np.select(conditions, choices)

# Print the DataFrame to see the new column
df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Calculate the total before tax
total_before_tax = df['line_subtotal'] + df['shipping_price']

# Calculate the tax amount as 9.25% of the total before tax
tax_amount = total_before_tax * 0.0925

# Calculate the total price by adding the total before tax and the tax amount
df['line_price'] = total_before_tax + tax_amount

# Round the line price to the nearest penny
df['line_price'] = df['line_price'].round(2)

# Print the DataFrame to see the new column
df[['line_subtotal', 'shipping_price', 'line_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

# Calculate the cost for each line
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']

# Print the DataFrame to see the new column
df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['line_price'] - df['line_cost']

# Display the DataFrame to see the new column along with other relevant information
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

# Identify the top 5 clients by quantity
client_counts = df.groupby('client_id').size().sort_values(ascending=False)
top_5_clients = client_counts.head(5).index.tolist()

# Calculate the total spent by each of the top 5 clients
client_spending = df.groupby('client_id')['line_price'].sum()
top_clients_spending = client_spending.loc[top_5_clients]

# Print the results
for client_id, total_spent in top_clients_spending.items():
    print(f"{client_id}: ${total_spent:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Identify the top 5 clients by quantity
top_5_clients = client_counts.head(5).index.tolist()

# Calculate the totals for each of the top 5 clients
client_totals = df.groupby('client_id')[['qty', 'shipping_price', 'line_price', 'line_cost', 'line_profit']].sum()
top_clients_totals = client_totals.loc[top_5_clients]

# Sort by total profit
top_clients_totals_sorted = top_clients_totals.sort_values('line_profit', ascending=False)

# Display the results
top_clients_totals_sorted
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Rename the columns
top_clients_totals_sorted.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Units',
    'shipping_price': 'Shipping',
    'line_price': 'Total Revenue',
    'line_cost': 'Total Cost',
    'line_profit': 'Total Profit'
}, inplace=True)

# Convert to millions of dollars
currency_columns = ['Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']
for col in currency_columns:
    top_clients_totals_sorted[col] = top_clients_totals_sorted[col] / 1_000_000

# Format the currency columns
for col in currency_columns:
    top_clients_totals_sorted[col] = top_clients_totals_sorted[col].map('${:,.2f}M'.format)
    
# Display the results
top_clients_totals_sorted
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
sorted_df = top_clients_totals_sorted.sort_values(by='Total Profit', ascending=False)

# Display the sorted DataFrame
sorted_df
```
